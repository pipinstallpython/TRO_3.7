##### У нас есть изображение пейзажа, на котором присутствуют деревья, небо и трава, и нам нужно изменить цвет травы на изображении, чтобы сделать ее более яркой и насыщенной.

###### Для решения задачи:

Загрузим изображение пейзажа и преобразуем его в нужное цветовое пространство HSV.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('kartinki-vikingi-3.jpg')
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

Найдем диапазон цветов, соответствующих траве на изображении, используя цветовые фильтры и пороговую фильтрацию.

```python
lower_green = np.array([50, 50, 50])
upper_green = np.array([70, 255, 255])
```

Изменим цвет травы на изображении, путем изменения значений цветовых каналов, увеличивая насыщенность и яркость зеленого цвета.

```python
mask = cv2.inRange(hsv, lower_green, upper_green)
```

```python
hsv[:, :, 1] += 50
hsv[:, :, 2] += 50
```

```python
result = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
```

Визуализируем результаты обработки с помощью библиотеки matplotlib.

```python
plt.subplot(121), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB)), plt.title('Processed')
plt.xticks([]), plt.yticks([])
plt.show()
```
